```python
from rdkit.Chem import AllChem as Chem
```

```python
smiles = 'COC(=O)c1c[nH]c2cc(OC(C)C)c(OC(C)C)cc2c1=O'
mol = Chem.MolFromSmiles(smiles)
print(mol)
```

```python
mol
```

```python
# Convert a RDKit molecule to SMILES

smi = Chem.MolToSmiles(mol)
smi
```

```python
# Convert a RDKit molecule to InchiKey.


Chem.MolToInchiKey(mol)
```

```python
#Convert a RDKit molecule to coordinative representation (which can be stored in .sdf file)


mol_block = Chem.MolToMolBlock(mol)
print(mol_block)
```

```python
count = 0
for line in 'smiles.csv':
    count +=1
    
print('Number of smiles entries', count)
```

```python
%time
import time
from psikit import Psikit
energy = []
time_data = []
for i in range(count-1):
    print("Molecule number:", i+1)
    time_init = time.time()
    pk=Psikit(debug=True, threads=1, memory=12)
    pk.read_from_smiles(data['smiles'][i])
    opt=pk.optimize()
    energy.append(opt)
    print("optimized SCF energy", opt)
    time_final = time.time()
    time_data.append(time_final-time_init)
```

```python
print(time_data)
```

```python
import numpy as np
esol_data["Energy (Hartree)"] = np.array(energy)
```

```python
esol_data
```

```python
esol_data["CPU time"] = np.array(time_data)
```
