```python
import numpy as np
import matplotlib.pyplot as plt
```

# Plots and data for MD simulation

```python
NPT_path = "./../Outputs/NPT-1.ener"
NVT_path = "./../Outputs/NVT-1.ener"
```

```python
npt_data = np.loadtxt(NPT_path)
nvt_data = np.loadtxt(NVT_path)
```

```python
npt_time = npt_data[:,1]
nvt_time = nvt_data[:,1]
npt_Ep = npt_data[:,4]
nvt_Ep = nvt_data[:,4]
```

```python
print(npt_time.shape)
print(npt_Ep.shape)
```

```python
plt.plot(npt_time, npt_Ep, label="NPT-simulation data")
plt.plot(nvt_time, nvt_Ep, label="NVT-simulatin data")
plt.legend()
plt.xlabel("Time (fs)")
plt.ylabel("Potential energy (a.u.)")
plt.title ("Potential energy plots for MD simulation \n NPT and NVT ensemble", fontsize = 15)
plt.savefig('./../Figures/potential_energy_MD.png')
plt.show()
```
