```python
import numpy as np
import matplotlib.pyplot as plt
```

# Reaction Coordinates vs Time plots

```python
reac_coord_path1 = "./../Outputs/monitor/MONITOR-COLVAR.metadynLog"
reac_coord_data1 = np.loadtxt(reac_coord_path1)
```

```python
rc_time1 = reac_coord_data1[:,0]
rc_reac_coord1 = reac_coord_data1[:,1]
```

```python
plt.plot(rc_time1, rc_reac_coord1, label="reaction-coords vs time")
plt.xlabel("Time (fs)")
plt.ylabel("Reaction Coordinate (d$_1$+coeff*d$_2$)")
plt.legend()
plt.savefig('./../Figures/reac_coord_vs_time-monitor_dir.png')
plt.show()
```

# QM/MM and Metadynamics Simulation

```python
data1_path = "./../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog"
data1 = np.loadtxt(data1_path)
```

```python
data1_CV = data1[:,1]
data1_time = data1[:,0]
```

```python
plt.plot(data1_time, data1_CV, label = "Standard QM/MM from 'Monitor' directory")
plt.legend()
plt.suptitle("Standard QM/MM simulation ", fontsize = 15)
plt.title("Hill Height = 1.5 kJ/mol, Freq = 100, Width = 0.5")
plt.xlabel("Timesteps")
plt.ylabel("Relative Energy (in kcal / mol)")
plt.savefig('./../Figures/reac_coord_QM-MM_height-1.5_freq-100_width-0.5.png')
plt.show()
```

```python
data2_path = "./../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog"
data2 = np.loadtxt(data2_path)
```

```python
data2_CV = data2[:,1]
data2_time = data2[:,0]
```

```python
plt.plot(data2_time, data2_CV, label = "Standard QM/MM from 'metad1_initmonitor' directory")
plt.legend(bbox_to_anchor = (1.04,1))
plt.suptitle("QM/MM plus MetaD simulation ", fontsize = 15)
plt.title("Hill Height = 1.5 kJ/mol, Freq = 100, Width = 0.5")
plt.xlabel("Timesteps")
plt.ylabel("Relative Energy (in kcal / mol)")
plt.savefig('./../Figures/reac_coord_QM-MM_MetaD_height-1.5_freq-100_width-0.5.png')
plt.show()
```
