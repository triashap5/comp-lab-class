# 6. finding end-to-end distance

```python
traj = md.load_xtc('1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='1hz3_T310.start.nowater.gro')
```

```python
def end_end(traj):
    """Compute the end to end distance of a trajectory"""
    # get the positions of the first and last atoms
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atoms
    dist = np.linalg.norm(first - last, axis=1)
    return dist
```

```python
# compute the end to end distance
dist = end_end(traj)
print(dist)
```

```python
# print the mean and standard deviation
print('mean = ', np.mean(dist))
print('std = ', np.std(dist))
```

```python
# plot the distribution
import matplotlib.pyplot as plt
plt.hist(dist, bins=20)
plt.xlabel('end to end distance (nm)')
plt.ylabel('frequency')
```

## 7. Radius of Gyration

```python
r_o_gyr = md.compute_rg(traj)
print(r_o_gyr)
```

```python
print('mean = ', np.mean(r_o_gyr))
print('std = ', np.std(r_o_gyr))
```

```python
plt.hist(r_o_gyr, bins=20)
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('frequency')
```

```python
# define the function to compute the distance and radius
# compute the distance between the first and last atom
def end_end(traj):
    # get the first and last atom
    first = traj.xyz[:,0,:]
    last = traj.xyz[:,-1,:]
    # compute the distance between the first and last atom
    dist = np.sqrt(np.sum((first-last)**2, axis=1))
    return dist
# compute the radius of gyration
def radius_of_gyration(traj):
    rg = md.compute_rg(traj)
    return rg
```

```python
fig, ax1 = plt.subplots()
dist = end_end(traj)
rg = radius_of_gyration(traj)
plt.title('End-End distance and Radius of Gyration vs Time')
ax1.plot(traj.time/1000, dist, color='blue', linestyle = 'solid', linewidth = 2)
ax1.set_xlabel('Time (ns)')
ax1.set_ylabel('End-End Distance (nm)', color = 'red')
ax1.tick_params('y', colors = 'black')

ax2 = ax1.twinx()
ax2.plot(traj.time/1000, rg, color = 'Magenta', linestyle = 'dashed', linewidth = 2)
ax2.set_ylabel('Radius of Gyration (nm)', color = 'Magenta')
ax2.tick_params('y', colors = 'Magenta')

plt.show()
```
