```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import pandas as pd
import itertools as itt
```

## 1. Plot for Energy minimization

```python
# Plot of Potential Energies at each minimization step
pot_en = np.loadtxt('./potential.xvg', skiprows = 24)
```

```python
time = pot_en[:,0]
E_pot = pot_en[:,1]
```

```python
plt.xlabel ('Energy Minimization Step')
plt.ylabel ('Potential Energy (kJ/mol)')
plt.plot(time,E_pot)
plt.title('Energy Minimization by Steepest Descent Method')
plt.show()
plt.savefig('./Energy_minimization.png')
```

## Plot for RMSD relative to the crystal structure of 1AKI

```python
rmsd_eq = np.loadtxt('./rmsd.xvg', skiprows = 18)
rmsd_crys = np.loadtxt('./rmsd_xtal.xvg', skiprows = 18)
```

```python
eq_time = rmsd_eq[:,0]
eq_rmsd = rmsd_eq[:,1]
```

```python
crys_time = rmsd_crys[:,0]
crys_rmsd = rmsd_crys[:,1]
```

```python
plt.figure()
plt.plot(eq_time, eq_rmsd, crys_time, crys_rmsd)
plt.xlim([0.0, 1.0])
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (ns)')
plt.legend(['Ref: Equilibriated', 'Ref: Crystal'])
plt.title('RMSD for Protein backbone')
plt.show()
plt.savefig('./RMSD.png')
```
