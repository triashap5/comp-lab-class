## Ions (NaCl) in water

```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import pandas as pd
import itertools 
```

## MSD Calculation

```python
msd_Na = np.loadtxt('./Na.xvg', skiprows = 20)
msd_Cl = np.loadtxt('./Cl.xvg', skiprows = 20)
msd_Oxy = np.loadtxt('./Oxy.xvg', skiprows = 20)
```

```python
Na_time = msd_Na[:,0]
Na_msd = msd_Na[:,1]
```

```python
Cl_time = msd_Cl[:,0]
Cl_msd = msd_Cl[:,1]
```

```python
Oxy_time = msd_Oxy[:,0]
Oxy_msd = msd_Oxy[:,1]
```

```python
plt.figure()
plt.plot(Na_time, Na_msd, Cl_time, Cl_msd, Oxy_time, Oxy_msd)
plt.xlabel('Time (ns)')
plt.ylabel('MSD (ns)')
plt.legend(['Ref: Na', 'Ref: Cl', 'Ref: Oxygen'])
plt.title('MSD for NaCl in water system')
plt.show()
plt.savefig('../Figures/MSD.png')
```
