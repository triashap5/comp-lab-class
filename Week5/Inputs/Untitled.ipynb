```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
import mdtraj as md
import pandas as pd
```

```python
pot_en_300 = np.loadtxt('./T300/Epot.xvg', skiprows = 24)
pot_en_450 = np.loadtxt('./T450/Epot.xvg', skiprows = 24)
pot_en_600 = np.loadtxt('./T600/Epot.xvg', skiprows = 24)
pot_en_1000 = np.loadtxt('./T1000/Epot.xvg', skiprows = 24)
```

```python
E_pot_300 = pot_en_300[:,1]
E_pot_450 = pot_en_450[:,1]
E_pot_600 = pot_en_600[:,1]
E_pot_1000 = pot_en_1000[:,1]
```

```python
plt.hist(E_pot_300, bins = 50, color = 'w', edgecolor = 'red', linewidth = 0.8)
plt.hist(E_pot_450, bins = 50, color = 'w', edgecolor = 'orange', linewidth = 0.8)
plt.hist(E_pot_600, bins = 50, color = 'w', edgecolor = 'green', linewidth = 0.8)
plt.hist(E_pot_1000, bins = 50, color = 'w', edgecolor = 'blue', linewidth = 0.8)
plt.legend(['300K', '450K', '600K', '1000K'])
plt.xlabel('Potential Energy')
plt.ylabel('Frequency')
plt.title('Potential Energy Histogram')
plt.show()
plt.savefig('./Figures/histogram_final.png')
```
