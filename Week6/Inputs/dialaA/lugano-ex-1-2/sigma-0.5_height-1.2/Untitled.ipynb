```python
fes = np.loadtxt('./fes.dat', skiprows = 5)
phi_fes = fes[:,0]
energy_fes = fes[:,1]
```

```python
for i in range(0,100):
    fes_i= np.loadtxt('./fes_'+str(i)+'.dat', skiprows = 5)
    phi_fes_i = fes_i[:,0]
    energy_fes_i = fes_i[:,1]
    plt.plot(phi_fes_i, energy_fes_i, label = str((i+1)/10)+"ns")
    
plt.xlabel('$\phi$ (rad)')
plt.ylabel('Energy (kJ/mol)')
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```
