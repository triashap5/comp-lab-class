```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
import mdtraj as md
import pandas as pd
from numpy import random
```

# Checking convergence rate with change in sigma values

## Sigma = 0.3, Height = 1.2

```python
for i in range(0,100):
    fes_i= np.loadtxt('./sigma-0.3_height-1.2/fes_'+str(i)+'.dat', skiprows = 5)
    phi_fes_i = fes_i[:,0]
    energy_fes_i = fes_i[:,1]
    plt.plot(phi_fes_i, energy_fes_i, label = str((i+1)/10)+"ns")
    
plt.xlabel('$\phi$ (rad)')
plt.ylabel('Energy (kJ/mol)')
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.title('dialaB-sigma-0.3-height-1.2-bf-10')
plt.savefig('../../../figures/dialaB-sigma-0.3-height-1.2-bf-10.png')
plt.show()
```
