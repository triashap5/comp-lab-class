```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
import mdtraj as md
import pandas as pd
from numpy import random
```

# Checking convergence rate with change in Bias Factor

## Sigma = 0.5, Height = 1.2, Biasfactor = 5

```python
for i in range(0,100):
    fes_i= np.loadtxt('./sigma-0.5-height-1.2-biasf-5/fes_'+str(i)+'.dat', skiprows = 5)
    phi_fes_i = fes_i[:,0]
    energy_fes_i = fes_i[:,1]
    plt.plot(phi_fes_i, energy_fes_i, label = str((i+1)/10)+"ns")
    
plt.xlabel('$\phi$ (rad)')
plt.ylabel('Energy (kJ/mol)')
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.title('dialaB-sigma-0.5-height-1.2-bf-5')
plt.savefig('../../../figures/dialaB-sigma-0.5-height-1.2-bf-5.png')
plt.show()
```

## Plotting final free energy for different bias factors

```python
fes_bf_5 = np.loadtxt('./sigma-0.5-height-1.2-biasf-5/fes.dat', skiprows = 5)
fes_bf_10 = np.loadtxt('./sigma-0.5-height-1.2-biasf-10/fes.dat', skiprows = 5)
fes_bf_15 = np.loadtxt('./sigma-0.5-height-1.2-biasf-15/fes.dat', skiprows = 5)

phi_bf5 = fes_bf_5[:,0]
phi_bf10 = fes_bf_10[:,0]
phi_bf15 = fes_bf_15[:,0]

fes_5 = fes_bf_5[:,1]
fes_10 = fes_bf_10[:,1]
fes_15 = fes_bf_15[:,1]
```
