```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import numpy as np
import pandas as pd
```

```python
fes = pd.read_csv('./fes.dat', sep='\s+', skiprows=9, header=None)
fes.columns = ['phi', 'psi', 'free ene', 'der_phi', 'der_psi']
fes
```

```python
phi=fes['phi'].to_numpy()
psi=fes['psi'].to_numpy()
ene=fes['free ene'].to_numpy()
```

```python
phi_new=phi[:33]
psi_new=psi[::33]
ene_new=ene.reshape(18,33)
```

```python
plt.figure()
plt.imshow(ene_new, origin='lower', interpolation='nearest', extent=[phi_new[0], phi_new[-1], psi_new[0], psi_new[-1]], cmap='viridis')
plt.xlabel('$\phi$')
plt.ylabel('$\psi$')
plt.title('Free Energy Surface-dialaB-2D')
plt.colorbar(label='Free Energy (kJ/mol)')
plt.savefig('FES_part_6.png')
plt.show()
```

```python
plt.contourf(phi_new, psi_new, ene_new, 20)
plt.xlabel('$\phi$')
plt.ylabel('$\psi$')
plt.title('Free Energy Surface-dialaB-2D')
plt.colorbar(label='Free Energy (kJ/mol)')
plt.savefig('../../../figures/dialaB-fes-metad2D.png')
plt.show()
```
