```python
import numpy as np
import MDAnalysis as mda
import matplotlib.pyplot as plt
import MDAnalysis.analysis.msd as msd
import MDAnalysis.analysis.rdf as rdf
from MDAnalysis import Universe
```

```python
#Loading topology
topol = '../Data/equil/kalj_T2_n360_v300_10000_1.lammpstrj'

# paths for trajectory data 
trajpath_T1p5 = "../Data/prod/kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd"           #T=1.5
trajpath_T1p0 = "../Data/prod/kalj_T1_n360_v300_prod_50000_1.unwrap.dcd"             #T=1.0
trajpath_T0p9 = "../Data/prod/kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd"          #T=0.9
trajpath_T0p8 = "../Data/prod/kalj_T0.8_n360_v300_prod_150000_1.unwrap.dcd"          #T=0.8
trajpath_T0p7 = "../Data/prod/kalj_T0.7_n360_v300_prod_180000_1.unwrap.dcd"          #T=0.7
trajpath_T0p65 = "../Data/prod/kalj_T0.65_n360_v300_prod_350000_1.unwrap.dcd"        #T=0.65
trajpath_T0p6 = "../Data/prod/kalj_T0.6_n360_v300_prod_1000000_1.unwrap.dcd"         #T=0.6
trajpath_T0p55 = "../Data/prod/kalj_T0.55_n360_v300_prod_3000000_1.unwrap.dcd"       #T=0.55
trajpath_T0p5 = "../Data/prod/kalj_T0.5_n360_v300_prod_10000000_1.unwrap.dcd"        #T=0.5
trajpath_T0p475 = "../Data/prod/kalj_T0.475_n360_v300_prod_30000000_1.unwrap.dcd"    #T=0.475
trajpath_T0p45 = "../Data/prod/kalj_T0.45_n360_v300_prod_100000000_1.unwrap.dcd"     #T=0.45




# load trajectory files with mdanalysis universe
traj_T1p5 = mda.Universe(topol, trajpath_T1p5, topology_format="LAMMPSDUMP")
traj_T1 = mda.Universe(topol, trajpath_T1p0, topology_format='LAMMPSDUMP')
traj_T0p9 = mda.Universe(topol, trajpath_T0p9, topology_format='LAMMPSDUMP')
traj_T0p8 = mda.Universe(topol, trajpath_T0p8, topology_format='LAMMPSDUMP')
traj_T0p7 = mda.Universe(topol, trajpath_T0p7, topology_format='LAMMPSDUMP')
traj_T0p65 = mda.Universe(topol, trajpath_T0p65, topology_format='LAMMPSDUMP')
traj_T0p6 = mda.Universe(topol, trajpath_T0p6, topology_format='LAMMPSDUMP')
traj_T0p55 = mda.Universe(topol, trajpath_T0p55, topology_format='LAMMPSDUMP')
traj_T0p5 = mda.Universe(topol, trajpath_T0p5, topology_format='LAMMPSDUMP')
traj_T0p475 = mda.Universe(topol, trajpath_T0p475, topology_format='LAMMPSDUMP')
traj_T0p45 = mda.Universe(topol, trajpath_T0p45, topology_format='LAMMPSDUMP')
```

```python
#RDF Calculations



T = [1.5, 1, 0.9, 0.8, 0.7, 0.65, 0.6, 0.55, 0.5, 0.475, 0.45]


traj_list = [traj_T1p5, traj_T1, traj_T0p9, traj_T0p8, traj_T0p7, traj_T0p65, traj_T0p6, traj_T0p55, traj_T0p5, 
            traj_T0p475, traj_T0p45]



rdf_values = np.empty((len(T),3), dtype = object)

    
for i in range (len(T)):
    atype_1 = traj_list[i].select_atoms("type 1")
    atype_2 = traj_list[i].select_atoms("type 2")
    
    rdf_values[i,0] = rdf.InterRDF(atype_1, atype_1, nbins=100, range = (0, 10))
    rdf_values[i,1] = rdf.InterRDF(atype_2, atype_2, nbins=100, range = (0, 10))
    rdf_values[i,2] = rdf.InterRDF(atype_1, atype_2, nbins=100, range = (0, 10))
    
    
    rdf_values[i,0].run()
    rdf_values[i,1].run()
    rdf_values[i,2].run()
```

```python
# PLOTS FOR RDFs (atom type 1 with atom type 1)


colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB', '#332288', '#88CCEE', '#44AA99', '#117733']
linewidths = np.linspace(2, 1, 11)



for i in range(len(T)):
    plt.plot(rdf_values[i,0].bins, rdf_values[i,0].rdf, color=colors[i], linewidth=linewidths[i], label='T='+str(T[i]), alpha=0.8)
    
plt.legend()
plt.title('RDF values between atom types 1 and 1 (g$_{11}$(r)) \n 3D KALJ glass-forming liquid', fontsize=12)
plt.xlabel('r (Angstroms)')
plt.ylabel('g(r)')
plt.savefig('../figures/rdf-1-1-KALJ.pdf')
```

```python
#PLOTS FOR RDFs (atom type 2 with atom type 2)


colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB', '#332288', '#88CCEE', '#44AA99', '#117733']
linewidths = np.linspace(2, 1, 11)



for i in range(len(T)):
    plt.plot(rdf_values[i,1].bins, rdf_values[i,1].rdf, color=colors[i], linewidth=linewidths[i], label='T='+str(T[i]), alpha=0.8)
    
plt.legend()
plt.xlabel('r (Angstroms)')
plt.ylabel('g(r)')
plt.title('RDF values between atom types 2 and 2 (g$_{22}$(r)) \n 3D KALJ glass-forming liquid', fontsize=12)
plt.savefig('../figures/rdf-2-2-KALJ.pdf')
```

```python
# PLOTS FOR RDFs (atom type 1 with atom type 2)


colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB', '#332288', '#88CCEE', '#44AA99', '#117733']
linewidths = np.linspace(2, 1, 11)



for i in range(len(T)):
    plt.plot(rdf_values[i,2].bins, rdf_values[i,2].rdf, color=colors[i], linewidth=linewidths[i], label='T='+str(T[i]), alpha=0.8)
    
plt.legend()
plt.xlabel('r (Angstroms)')
plt.ylabel('g(r)')
plt.title('RDF values between atom types 1 and 2 (g$_{12}$(r)) \n 3D KALJ glass-forming liquid', fontsize=12)
plt.savefig('../figures/rdf-1-2-KALJ.pdf')
```

```python
# MSD data generation

#Temperature values
T = [1.5, 1, 0.9, 0.8, 0.7, 0.65, 0.6, 0.55, 0.5, 0.475, 0.45]

#define trajectory list
traj_list = [traj_T1p5, traj_T1, traj_T0p9, traj_T0p8, traj_T0p7, traj_T0p65, traj_T0p6, traj_T0p55, traj_T0p5, traj_T0p475, traj_T0p45]
#create empty list to store rdf values

msd_values = np.empty(len(T), dtype=object)
#select atom type 1
sel = 'type 1'

# calculate MSDs
for i in range (len(T)):
    
    # compute msd
    msd_values[i] = msd.EinsteinMSD(traj_list[i], select=sel, msd_type='xyz', fft=False)
    # run
    msd_values[i].run()
```

```python
#Plots for mean squared displacement

T = [1.5, 1, 0.9, 0.8, 0.7, 0.65, 0.6, 0.55, 0.5, 0.475, 0.45]

for i in range(len(T)):
    
    plt.loglog(msd_values[i].times, msd_values[i].results.timeseries, color=colors[i], linewidth=linewidths[i], label='T'+str(T[i]), alpha=0.8)

plt.xlabel('time (ps)')
plt.ylabel('MSD (Å$^2$)')
plt.legend(loc='upper left', frameon=True)
plt.tick_params(left = False)
plt.tick_params(bottom = False)
plt.xlim(0.00003, 5000)

plt.title('Mean Squared Displacement Values <$\Delta r^2$> \n 3D KALJ glass-forming liquid')
plt.savefig('../figures/msd-KALJ.pdf')
```

```python
#calculation for diffusion times. 
#Find the time where MSD = 1 at each temperature. This is the time where the particle would diffuse 1 diameter on average.

T = [1.5, 1, 0.9, 0.8, 0.7, 0.65, 0.6, 0.55, 0.5, 0.475, 0.45]

diffusion_times = []
for i in range(len(T)):
    # try find the index of the first time where MSD > 1
    try :
        index = np.where(msd_values[i].results.timeseries > 1)[0][0]
    # if not found, set index to -1
    except IndexError:
        index = -1
    # append the diffusion time to the list
    diffusion_times.append(msd_values[i].times[index])
```

```python
#Plots for diffusion times

plt.loglog(1/np.array(T, dtype=float), diffusion_times,'--o', color='#0077BB', alpha=0.8, label='Diffusion time')
plt.xlabel('1/T (K$^{-1}$)')
plt.ylabel('Diffusion time (ps)')
plt.xlim(0.65, 2.3)
plt.title('Diffusion Times \n 3D KALJ glass-forming liquid')
plt.legend()
plt.savefig('../figures/diffusion_times-KALJ.pdf')
```
